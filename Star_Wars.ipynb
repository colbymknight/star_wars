# Star Wars Survey
## Overview

```python
import pandas as pd
import numpy as np
star_wars = pd.read_csv('star_wars.csv', encoding='ISO=8859-1')
```

```python
star_wars.head(10)
```

## Cleaning/Mapping Yes/No Columns

```python
star_wars.columns
```

```python
yes_no = {"Yes": True, "No": False}

for col in [
    "Have you seen any of the 6 films in the Star Wars franchise?",
    "Do you consider yourself to be a fan of the Star Wars film franchise?"
    ]:
    star_wars[col] = star_wars[col].map(yes_no)

star_wars.head()
```

```python
star_wars.head()
```

```python
movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True,
    np.nan: False,
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
```

```python
star_wars = star_wars.rename(columns={
    'Which of the following Star Wars films have you seen? Please select all that apply.': 'seen_ep1',
    'Unnamed: 4': 'seen_ep2',
    'Unnamed: 5': 'seen_ep3',
    'Unnamed: 6': 'seen_ep4',
    'Unnamed: 7': 'seen_ep5',
    'Unnamed: 8': 'seen_ep6',
})
```

```python
star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars = star_wars.rename(columns={
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': 'rank_1',
    'Unnamed: 10': 'rank_2',
    'Unnamed: 11': 'rank_3',
    'Unnamed: 12': 'rank_4',
    'Unnamed: 13': 'rank_5',
    'Unnamed: 14': 'rank_6',
})
```

## Find the Highest-Ranked Movie

```python
ranks = star_wars[star_wars.columns[9:15]].mean()
print(ranks)
```

```python
%matplotlib inline
ranks.plot.bar()
```

Remembering that a lower ranking is better, the average rank of each movie from best to worst is:
* Episode 5
* Episode 6
* Episode 4
* Episode 1
* Episode 2
* Episode 3

## Finding the Most Viewed Movie

```python
seen = star_wars[star_wars.columns[3:9]].sum()
print(seen)
```

```python
seen.plot.bar()
```

Below are the movies ranked from most to least seen:
* Episode 5
* Episode 6
* Episode 1
* Episode 4
* Episode 2
* Episode 3

```python
males = star_wars[star_wars['Gender'] == 'Male']
females = star_wars[star_wars['Gender'] == 'Female']
```

```python
male_seen = males[males.columns[3:9]].sum()
print(male_seen)
```

```python
male_seen.plot.bar()
```

It would appear that being male has little to no effect on the view rates of each movie, so one would conclude the same of the female set.

```python
male_ranks = males[males.columns[9:15]].mean()
male_ranks.plot.bar()
```

While viewership remains the same among males as both male and female, there is definitely an increased preference among males for the origial trilogy movies.

```python
female_ranks = females[females.columns[9:15]].mean()
print(female_ranks)
```

```python
female_ranks.plot.bar()
```

As expected, there is less of an obvious divide between the original trilogy and the prequel trilogy among female viewers
